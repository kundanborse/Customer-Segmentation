# Creating Clusters:

# First, we will create a cluster for ‘Age’ and ‘Spending Score’ columns.

```python
from sklearn.cluster import KMeans
x = data.loc[:,['Age','Spending Score (1-100)']].values
```

```python
# to plot elbow graph
wcss = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k, init='k-means++')
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()    
```

```python
kmeans = KMeans(n_clusters = 4)
y_kmeans = kmeans.fit_predict(x)

print(y_kmeans)
print(kmeans.cluster_centers_)
```

```python
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], s = 80, c = 'red', label = 'Cluster 1')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], s = 80, c = 'blue', label = 'Cluster 2')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1], s = 80, c = 'green', label = 'Cluster 3')
plt.scatter(x[y_kmeans == 3, 0], x[y_kmeans == 3, 1], s = 80, c = 'cyan', label = 'Cluster 4')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], color = 'black')
plt.title('Cluster of Customers')
plt.xlabel('Age')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```

```python
from sklearn.cluster import KMeans
x1 = data.loc[:,['Annual Income (k$)','Spending Score (1-100)']].values
```
